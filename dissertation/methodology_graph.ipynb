```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Data for pyramid levels
levels = [
    {"label": "20,540\nSingle Extracted Images", "value": 20540},
    {"label": "3,563\nSame-Image Clusters", "value": 3563},
    {"label": "168\nContent Keywords", "value": 168}
]

# Plot setup
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_xlim(0, 10)
ax.set_ylim(0, 6)
ax.axis('off')

# Pyramid base width
base_width = 9
height_step = 2

# Draw the pyramid layers
for i, level in enumerate(levels):
    width = base_width * ((len(levels) - i) / len(levels))
    left = (10 - width) / 2
    bottom = i * height_step
    rect = patches.Rectangle((left, bottom), width, height_step,
                             linewidth=1, edgecolor='black', facecolor='#cfe2f3')
    ax.add_patch(rect)
    ax.text(5, bottom + height_step / 2, level["label"],
            ha='center', va='center', fontsize=12, weight='bold')

# Title
# plt.title("From Extracted Images to Content Keywords", fontsize=14, weight='bold')
plt.tight_layout()
plt.savefig("/Users/nogashlomi/projects/nog_thesis/figures/plots/methodology/pyramid.png")
plt.show()
```
