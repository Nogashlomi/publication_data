```python
import shutil
from pathlib import Path

# --- List of SAC_SIL files to copy ---
sac_sil_files = [
    "SAC_SIL_00090", "SAC_SIL_00195", "SAC_SIL_00250", "SAC_SIL_00778",
    "SAC_SIL_01161", "SAC_SIL_01195", "SAC_SIL_01519", "SAC_SIL_02055",
    "SAC_SIL_02190", "SAC_SIL_02227", "SAC_SIL_02304", "SAC_SIL_02308",
    "SAC_SIL_02326", "SAC_SIL_02469", "SAC_SIL_02470", "SAC_SIL_02528",
    "SAC_SIL_02572", "SAC_SIL_03353", "SAC_SIL_03704", "SAC_SIL_03744",
    "SAC_SIL_03745", "SAC_SIL_03746", "SAC_SIL_03747", "SAC_SIL_03748",
    "SAC_SIL_03749", "SAC_SIL_03750", "SAC_SIL_03751"
]

# --- Paths ---
src_folder = Path("/Users/nogashlomi/Library/CloudStorage/GoogleDrive-noga.shlomi@gmail.com/My Drive/images_ck_flat")
dest_folder = Path.home() / "Desktop" / "selected_SAC_images"

# Create destination folder if it doesn't exist
dest_folder.mkdir(parents=True, exist_ok=True)

# --- Copy files ---
missing_files = []

for file_id in sac_sil_files:
    found = False
    for ext in ["jpg", "jpeg", "png"]:
        src_file = src_folder / f"{file_id}.{ext}"
        if src_file.exists():
            shutil.copy(src_file, dest_folder)
            print(f"✅ Copied: {src_file.name}")
            found = True
            break
    if not found:
        missing_files.append(file_id)

# --- Summary ---
if missing_files:
    print("\n⚠️ The following files were not found:")
    for f in missing_files:
        print(f)
else:
    print("\n🎉 All files copied successfully!")
```

```python
# ✅ Filter rows where cks == "CK_Astrological House"
astrological_houses = df[df['cks'] == "CK_Astrological Houses"].copy()

astrological_houses
```

```python
import matplotlib.pyplot as plt

# ✅ 1. Compute unique books per year_interval (filtered)
astro_books = astrological_houses.groupby('year_interval', observed=False)['book'].nunique()

# ✅ 2. Compute total unique books per year_interval (full df)
total_books = df.groupby('year_interval', observed=False)['book'].nunique()

# ✅ 3. Compute unique place names per year_interval (filtered)
places_per_interval = (
    astrological_houses.groupby('year_interval', observed=False)['place']
    .apply(lambda x: ', '.join(sorted(x.dropna().unique())))
)

# ✅ 4. Prepare data for plotting
intervals = astro_books.index.astype(str)

# ✅ 5. Plot both lines
plt.figure(figsize=(12, 7))

# Filtered CK line (Astrological Houses)
plt.plot(intervals, astro_books.values, marker='o', color='darkblue', label='Astrological Houses')

# Total books line (dashed)
plt.plot(total_books.index.astype(str), total_books.values, color='black', linestyle='--', label='All Books')

# ✅ 6. Annotate each blue dot with place names
for i, (x, y) in enumerate(zip(intervals, astro_books.values)):
    if y > 0:  # Only label intervals with books
        label = places_per_interval.iloc[i] if i < len(places_per_interval) else ''
        plt.text(x, y + 0.3, label, ha='center', va='bottom', fontsize=8, rotation=30)

# ✅ 7. Formatting
plt.title('Unique Books with "CK_Astrological House" per Year Interval')
plt.xlabel('Year Interval')
plt.ylabel('Number of Unique Books')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()

# ✅ 8. Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# ✅ 1. Total books per year_interval (full dataset)
all_books_count = df.groupby('year_interval', observed=False)['book'].nunique()

# ✅ 2. Total books per year_interval (filtered dataset)
filtered_books_count = filtered_df.groupby('year_interval', observed=False)['book'].nunique()

# ✅ 3. Use the full timeline from all_books_count
intervals = all_books_count.index.astype(str)

# ✅ 4. Create DataFrame for convenience
book_counts_df = pd.DataFrame({
    'All Books': all_books_count,
    'Filtered Books': filtered_books_count.reindex(all_books_count.index).fillna(0)
})

print("Number of Unique Books - Full vs Filtered:")
print(book_counts_df)

# ✅ 5. Collect place names from the full df for books in filtered_df, per interval
places_per_interval = {}

for interval, group in filtered_df.groupby('year_interval', observed=False):
    books_in_interval = group['book'].unique()
    places = df[df['book'].isin(books_in_interval)]['place'].dropna().unique()
    place_label = ', '.join(sorted(places))
    # Optionally shorten long labels
    if len(places) > 3:
        place_label = ', '.join(sorted(places)[:3]) + '...'
    places_per_interval[str(interval)] = place_label

# ✅ 6. Plotting
plt.figure(figsize=(10, 6))

# Full dataset (grey dashed)
plt.plot(intervals, book_counts_df['All Books'].values,
         label='All Books', marker='o', color='lightgrey', linestyle='--')

# Filtered dataset (black dashed)
plt.plot(intervals, book_counts_df['Filtered Books'].values,
         label='Filtered Books', marker='o', color='black', linestyle='--')

# ✅ 7. Add place labels near black dots (filtered books)
for x, y in zip(intervals, book_counts_df['Filtered Books'].values):
    if y > 0:  # Only label intervals with books
        label = places_per_interval.get(x, '')
        plt.text(x, y + 0.3, label, ha='center', va='bottom', fontsize=8, rotation=25)

# ✅ 8. Final touches
plt.title('Number of Unique Books Over Time: Filtered vs Full Dataset')
plt.xlabel('Year Interval')
plt.ylabel('Number of Unique Books')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()

# ✅ 9. Show
plt.show()
```
