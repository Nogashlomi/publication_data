# Text-parts Analysis for Motion of the Eighth Sphere
This notebook creates a scatter plot by text-part (`custom_identifier` and `part_or_adaption_label`) for the *Motion of the Eighth Sphere* keyword.


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the full image data
df = pd.read_csv('../../DATA/02_image_clusters/full_image_data_feb_25.csv', low_memory=False)

# Clean up nulls in essential columns
df_clean = df.dropna(subset=['year', 'part_or_adaption_label', 'custom_identifier']).copy()

# Convert custom_identifier to string for coloring
df_clean['custom_identifier_str'] = df_clean['custom_identifier'].astype(int).astype(str)

# Filter for the Motion of the Eighth Sphere
df_motion = df_clean[df_clean['cks'] == 'CK_Motion of the Eighth Sphere'].copy()
```

## Scatter Plot: Motion of the Eighth Sphere by Text Part
- X-axis is `year`
- Y-axis is `part_or_adaption_label` (Text Part)
- Dot is colored by `custom_id` number
- Hover data shows place, bid, and the exact keyword


```python
# Matplotlib scatter plot for Motion of the Eighth Sphere
plt.figure(figsize=(14, 8))

# Get unique Custom IDs to create a color mapping
unique_ids = df_motion['custom_identifier_str'].unique()
colors = plt.cm.tab20(np.linspace(0, 1, len(unique_ids)))
color_map = dict(zip(unique_ids, colors))

# Plot each group separately so they show up in the legend
for custom_id, group in df_motion.groupby('custom_identifier_str'):
    plt.scatter(
        group['year'], 
        group['part_or_adaption_label'], 
        color=color_map[custom_id],
        label=custom_id,
        s=80,      # Size of the dots
        alpha=0.7, # Transparency 
        edgecolors='w',
        linewidth=0.5
    )

# Formatting
plt.title('Motion of the Eighth Sphere by Year and Text Part', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Text Part / Adaption Label', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)

# Move the legend outside the plot box
plt.legend(title='Custom ID', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# SAVE THE IMAGE
# plt.savefig('/Users/nogashlomi/projects/nog_thesis/figures/corrections_round/eighth_sphere_1.png', dpi=300, bbox_inches='tight')

plt.show()
```
