```python
# target CKs list
target_cks = [
'CK_Lunar Eclipse',
'CK_Solar Eclipse',
'CK_Conical Shape of Planetary Shadow',
    'CK_Centrality of Earth', 
'CK_Sphericity of the Earth', 
'CK_Relation Between the Spheres of Water and Earth',
'CK_Lunar Phases',
'CK_Ephemerides of the Moon', 
'CK_Lunar Nodes',
'CK_Lunar Orbit'
]

# Use filtered_df instead of visual_df
places = df['place'].dropna().unique().tolist()

edges = []

for place in places:
    place_data = df[df['place'] == place]

    # Add edges only between place and target CK if place_data contains that CK
    for cks in target_cks:
        if (place_data['cks'] == cks).any():
            edges.append((place, cks))

# Unique nodes
places_nodes = list(set(edge[0] for edge in edges))
cks_nodes = list(set(edge[1] for edge in edges))

# Color palette and mapping (places by category)
ordered_categories = ['Large Center', 'Medium Center', 'Small Center', 'One Book Center']
existing_categories = df['place_category'].dropna().unique()
ordered_present_categories = [cat for cat in ordered_categories if cat in existing_categories]

color_palette = ['skyblue', 'salmon', 'lightgreen', 'khaki']
category_color_map = {cat: color for cat, color in zip(ordered_present_categories, color_palette)}

place_color_map = {}
for place in places_nodes:
    cat_series = df[df['place'] == place]['place_category'].dropna()
    category = cat_series.iloc[0] if not cat_series.empty else 'unknown'
    place_color_map[place] = category_color_map.get(category, 'gray')

# Build graph
G = nx.Graph()
G.add_nodes_from(places_nodes, bipartite=0)
G.add_nodes_from(cks_nodes, bipartite=1)
G.add_edges_from(edges)

# Node colors:
node_colors = []
for node in G.nodes():
    if node in places_nodes:
        node_colors.append(place_color_map.get(node, 'gray'))
    else:
        # target CK nodes colored grey
        node_colors.append('gray')

# Draw graph
plt.figure(figsize=(14, 10))
pos = nx.spring_layout(G, seed=42, k=0.8)  # Try 0.8, 1.0, or higher

nx.draw(
    G,
    pos,
    with_labels=True,
    node_color=node_colors,
    edge_color='gray',
    node_size=2500,
    font_size=9
)

# Legend
legend_handles = [
    mpatches.Patch(color=category_color_map[cat], label=cat)
    for cat in ordered_present_categories
]
legend_handles.append(mpatches.Patch(color='gray', label='Target CK'))

plt.legend(
    handles=legend_handles,
    title='Place Categories and Target CKs',
    loc='upper right',
    fontsize='small',
    title_fontsize='medium'
)
plt.title("Bipartite Graph: Places and Target CKs")

# plt.savefig('/Users/nogashlomi/projects/nog_thesis/figures/plots/source_analysis/4.3_cks_place_network.png', dpi=300)


plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
import matplotlib.patches as mpatches

# Define target CKs
target_cks = [
'CK_Lunar Eclipse',
'CK_Solar Eclipse',
'CK_Conical Shape of Planetary Shadow',
    'CK_Centrality of Earth', 
'CK_Sphericity of the Earth', 
'CK_Relation Between the Spheres of Water and Earth',
'CK_Lunar Phases',
'CK_Ephemerides of the Moon', 
'CK_Lunar Nodes',
'CK_Lunar Orbit'
]

# Prepare data
printers = df['printer'].dropna().unique().tolist()
publishers = df['publisher'].dropna().unique().tolist()

edges = []

# Add printer-CK edges
for printer in printers:
    printer_data = df[df['printer'] == printer]
    for cks in target_cks:
        if (printer_data['cks'] == cks).any():
            edges.append((f"PRINTER: {printer}", cks))

# Add publisher-CK edges
for publisher in publishers:
    publisher_data = df[df['publisher'] == publisher]
    for cks in target_cks:
        if (publisher_data['cks'] == cks).any():
            edges.append((f"PUBLISHER: {publisher}", cks))

# Build node sets
printer_nodes = list(set(edge[0] for edge in edges if edge[0].startswith("PRINTER:")))
publisher_nodes = list(set(edge[0] for edge in edges if edge[0].startswith("PUBLISHER:")))
cks_nodes = list(set(edge[1] for edge in edges))

# Color palette and mapping (place categories)
ordered_categories = ['Large Center', 'Medium Center', 'Small Center', 'One Book Center']
existing_categories = filtered_df['place_category'].dropna().unique()
ordered_present_categories = [cat for cat in ordered_categories if cat in existing_categories]
color_palette = ['skyblue', 'salmon', 'lightgreen', 'khaki']
category_color_map = {cat: color for cat, color in zip(ordered_present_categories, color_palette)}

# Assign colors to all named nodes (printers and publishers) by category
all_named_nodes = printer_nodes + publisher_nodes
combined_color_map = {}
for node in all_named_nodes:
    name = node.replace("PRINTER: ", "").replace("PUBLISHER: ", "")
    cat_series = filtered_df[
        (filtered_df['printer'] == name) | (filtered_df['publisher'] == name)
    ]['place_category'].dropna()
    category = cat_series.iloc[0] if not cat_series.empty else 'unknown'
    combined_color_map[node] = category_color_map.get(category, 'gray')

# Assign node colors
node_colors = []
for node in printer_nodes + publisher_nodes + cks_nodes:
    if node in combined_color_map:
        node_colors.append(combined_color_map[node])
    else:
        node_colors.append('gray')  # CKs

# Build graph
G = nx.Graph()
G.add_nodes_from(printer_nodes + publisher_nodes, bipartite=0)
G.add_nodes_from(cks_nodes, bipartite=1)
G.add_edges_from(edges)

# Draw graph
plt.figure(figsize=(16, 12))
pos = nx.spring_layout(G, seed=42)
nx.draw(
    G,
    pos,
    with_labels=False,
    node_color=node_colors,
    edge_color='gray',
    node_size=2500
)

# CK labels only
labels = {node: node.replace("CK_", "") for node in cks_nodes}
nx.draw_networkx_labels(G, pos, labels=labels, font_size=9)

# Legend
legend_handles = [
    mpatches.Patch(color=category_color_map[cat], label=f'Printer/Publisher ({cat})')
    for cat in ordered_present_categories
]
legend_handles.append(mpatches.Patch(color='gray', label='Target CK'))

plt.legend(
    handles=legend_handles,
    title='Node Types',
    loc='upper right',
    fontsize='small',
    title_fontsize='medium'
)

plt.title("Bipartite Graph: Printers, Publishers, and Target CKs")
# plt.tight_layout()

# plt.savefig('/Users/nogashlomi/projects/nog_thesis/figures/plots/source_analysis/4.3_cks_printers_publishers_network.png', dpi=300)


plt.show()
```

```python

# === Assumes this is already defined earlier ===
target_cks = [
'CK_Lunar Eclipse',
'CK_Solar Eclipse',
'CK_Conical Shape of Planetary Shadow'
]

# === Step 1: Visual tags ===
defined_columns = [
'3D', 
# 'decoration not faces',
'faces', 
'color orbs', 
'color shadow',
'multiple positions of an element', 
# 'full sphere circle',
'perspective from earth', 
# 'double images like comics', 
'human figures',
# 'lines conturing shadow', 
# 'lines for sight',
# 'double image two perspectives '
]

# === Step 2: Build edge list ===
edges = []
places = visual_df['place'].dropna().unique().tolist()

for place in places:
    place_data = visual_df[visual_df['place'] == place]

    # Add CKs only if in target_cks
    for cks in target_cks:
        if (place_data['cks'] == cks).any():
            edges.append((place, cks))

    # Add visual tag edges
    for tag in defined_columns:
        if (place_data[tag] == 'yes').any():
            edges.append((place, tag))

# Optional: inspect edge structure
# edges_df = pd.DataFrame(edges, columns=['place', 'tag'])
# print(edges_df)

# === Step 3: Unique node sets ===
places = list(set(edge[0] for edge in edges))
tags = list(set(edge[1] for edge in edges))

# === Step 4: Color by place category ===
ordered_categories = ['Large Center', 'Medium Center', 'Small Center', 'One Book Center']
existing_categories = visual_df['place_category'].dropna().unique()
ordered_present_categories = [cat for cat in ordered_categories if cat in existing_categories]

color_palette = ['skyblue', 'salmon', 'lightgreen', 'khaki']
category_color_map = {
    cat: color for cat, color in zip(ordered_present_categories, color_palette)
}

place_color_map = {}
for place in places:
    cat_series = visual_df[visual_df['place'] == place]['place_category'].dropna()
    category = cat_series.iloc[0] if not cat_series.empty else 'unknown'
    place_color_map[place] = category_color_map.get(category, 'gray')

# === Step 5: Build the graph ===
G = nx.Graph()
G.add_nodes_from(places, bipartite=0)
G.add_nodes_from(tags, bipartite=1)
G.add_edges_from(edges)

# === Step 6: Assign node colors ===
node_colors = [
    place_color_map[node] if node in places else 'lightgray'
    for node in G.nodes()
]

# === Step 7: Draw ===
plt.figure(figsize=(14, 10))
pos = nx.spring_layout(G, seed=42)
nx.draw(
    G,
    pos,
    with_labels=True,
    node_color=node_colors,
    edge_color='gray',
    node_size=2500,
    font_size=9
)

# === Step 8: Legend ===
legend_handles = [
    mpatches.Patch(color=category_color_map[cat], label=cat)
    for cat in ordered_present_categories
]
legend_handles.append(mpatches.Patch(color='lightgray', label='Tag (CK or Visual)'))

plt.legend(
    handles=legend_handles,
    title='Place Categories',
    loc='upper right',
    fontsize='small',
    title_fontsize='medium'
)
plt.title("Bipartite Graph: Places and Target CKs / Visual Tags")

# === Step 9: Save or show ===
# plt.savefig('/Users/nogashlomi/projects/nog_thesis/figures/plots/source_analysis/4.3_visual_cks_network.png', dpi=300)
plt.show()
```
