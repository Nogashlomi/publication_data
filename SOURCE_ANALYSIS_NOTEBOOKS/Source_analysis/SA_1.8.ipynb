```python
#correcting the tagging in df in a new csv

# Define general and subgroup CKs
general_ck = 'CK_Centrality of Earth'
subgroup_cks = [
    'CK_Negligible Dimensions of the Earth',
    'CK_Central Intersection of Eclipses',
    'CK_Visibility of Half of the Sky from the Central Earth'
]

# Get all image IDs tagged with the general CK
images_with_general = set(df[df['cks'] == general_ck]['images'].unique())

# Get all image IDs tagged with any of the subgroups
images_with_subgroups = set(df[df['cks'].isin(subgroup_cks)]['images'].unique())

# Identify images missing the general CK
images_to_fix = images_with_subgroups - images_with_general

# Create new rows for those images
new_rows = df[
    (df['images'].isin(images_to_fix)) & 
    (df['cks'].isin(subgroup_cks))
].drop_duplicates('images').copy()

# Assign the general CK
new_rows['cks'] = general_ck

# Append to the original DataFrame
df = pd.concat([df, new_rows], ignore_index=True)

print(f"✅ Added '{general_ck}' to {len(new_rows)} image entries.")

df.to_csv('../DATA/02_image_clusters/full_image_data_feb_25_corrected_centrality_earth.csv', index=False)

df = pd.read_csv('../DATA/02_image_clusters/full_image_data_feb_25_corrected_centrality_earth.csv')
```
